### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas
import sqlite3
import time
import os

os.remove("e10.db") if os.path.exists("e10.db") else None

conn = sqlite3.connect('e10.db')
cur = conn.cursor()
```

```python
prod1 = {'nome':'Geladeira','preco':2000.0, 'marca':'Samsung'}
prod2 = {'nome':'Televisão','preco':2500.0, 'marca':'LG'}
prod3 = {'nome':'Celular','preco':5600.0, 'marca':'Apple'}

prods = [prod1,prod2,prod3]
```

```python
cur.execute('CREATE TABLE IF NOT EXISTS produtos('\
    'nome varchar(50),'\
    'preco decimal,'\
    'marca varchar(50) )')
```

```python
class Produto():
    attr=['nome','preco','marca']
    def __init__(self,prod):
        self.nome = prod['nome']
        self.preco = prod['preco']
        self.marca = prod['marca']
```

```python
def insert(prod):
    cur.execute('INSERT INTO produtos(nome,preco,marca) VALUES(?,?,?)',\
        [prod.nome, prod.preco ,prod.marca])

for prod in prods:
    p = Produto(prod)
    insert(p)
    time.sleep(2)
    
```

```python
t = cur.execute('select * from produtos')
t.fetchall()
```
